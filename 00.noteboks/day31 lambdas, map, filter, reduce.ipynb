# Filters

```python
# filter() is an in-built function in python, based on some condition it filters the records

# syntax: filter(function, sequence)
```

```python
def is_even(num):
  if num % 2 == 0:
    return True
  else:
    return False
```

```python
nums = list(range(1, 11))
nums
```

```python
for num in nums:
  if is_even(num):
    print(num)
```

```python
filter(lambda num: num %2 == 0, nums)
```

```python
list(filter(lambda num: num %2 == 0, nums))
```

```python
tuple(filter(lambda num: num %2 == 0, range(1, 11)))
```

```python
list(filter(is_even, nums))
```

```python
# map(): it takes a sequence and it applies some functionality
# map(function, sequence)
```

```python
list(map(is_even, nums))
```

```python
list(map(lambda x: x+5, range(1, 11)))
```

```python
# reduce():
#   reduce(function, sequence)
#   it reduces a sequence of elements into a single element
```

```python
from functools import reduce
```

```python
nums = list(range(10, 51, 10))
nums # 10, 20, 30, 40, 50
```

```python
reduce(lambda x, y: x+y, [1, 2, 3, 4, 5])
```

```python
reduce(lambda x, y: x*y, [1, 2, 3, 4, 5])
```

```python
reduce(lambda x,y: x+y, nums)
```

# function aliasing

```python
def add(a, b):
  return a+b
```

```python
add(10, 20)
```

```python
a = add # assigning a function to a variable.
```

```python
a(10, 20)
```

```python
add(10, 20)
```

```python
id(add)
```

```python
id(a)
```
