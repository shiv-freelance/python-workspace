````python
class TooYoungException(Exception):
  def __init__(self, msg):
    self.msg = msg
````

````python
class Voter:
  def __init__(self, name, age):
    self.name = name
    self.age = age

  def __repr__(self): # string magic method --> this will give string representation of your object
    return f"Name of the voter is: {self.name} and age is {self.age}"
````

````python
def register_voter(name, age):
  if age < 18:
    raise TooYoungException("too young to cast a vote")
  v = Voter(name, age)
  return v
````
